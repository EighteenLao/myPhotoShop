```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
src = cv2.imread('C:\\Users\\USER\\Pictures\\Wallpaper\\164156003_x3.jpg')
```

# Day21: 油漆桶

```python
copy_img = src.copy()

seed = (0,0)
loDiff = 80
upDiff = 30
copy_img = src.copy()

h,w = copy_img.shape[:2]
mask = np.zeros([h+2,w+2],dtype=np.uint8)
cv2.floodFill(copy_img, mask, seed, (0, 125, 125), (loDiff, loDiff, loDiff), (upDiff, upDiff, upDiff), cv2.FLOODFILL_FIXED_RANGE) 


copy_img = cv2.cvtColor(copy_img, cv2.COLOR_BGR2RGB)
plt.imshow(copy_img)
plt.show()
```
