```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob
```

# Day17~19 圖像RGB邏輯運算、二值化、自適應圖像二值化

# Day17: 圖像RGB邏輯運算

```python
img1 = np.zeros((100,100,3),np.uint8)
img2 = np.zeros((100,100,3),np.uint8)

cv2.circle(img1,(30, 30), 30, (255,0, 255), -1)
cv2.circle(img2,(60, 60), 30, (255, 0, 0), -1)

image_or = cv2.bitwise_or(img1, img2, dst=None, mask=None)
image_and = cv2.bitwise_and(img1, img2, dst=None, mask=None) 
image_xor = cv2.bitwise_xor(img1, img2, dst=None, mask=None)
img1_not =cv2.bitwise_not(img1)
img2_not =cv2.bitwise_not(img2)

img = [img1,img2,image_or,image_and,image_xor,img1_not,img2_not]

for i in range(0,7):
    show_img = cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB)
    plt.imshow(show_img)
    plt.show()
```

# Day18: 圖像二值化運算

```python
#src = cv2.imread('C:\\Users\\USER\\Pictures\\Wallpaper\\thumb-1920-946628.png') #絕對路徑使用'\'，相對路徑使用'/'
src = cv2.imread('164156003_x3.jpg') #絕對路徑使用'\'，相對路徑使用'/'
```

```python
gray_img = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)

ret,img1 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
ret,img2 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY_INV)
ret,img3 = cv2.threshold(gray_img,127,255,cv2.THRESH_TRUNC)
ret,img4 = cv2.threshold(gray_img,127,255,cv2.THRESH_TOZERO)
ret,img5 = cv2.threshold(gray_img,127,255,cv2.THRESH_TOZERO_INV)
img = [img1,img2,img3,img4,img5]

for i in range(0,5):
    show_img = cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB)
    plt.imshow(show_img)
    plt.show()
```

# Day19: 自適應圖像二值化

```python
gray_img = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
blur_img =  cv2.GaussianBlur(gray_img,(5, 5), 0) 

#未濾波的二值化&自適應二值化
ret,img1 = cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
img2 = cv2.adaptiveThreshold(gray_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
#濾波的二值化&自適應二值化
ret,img3 = cv2.threshold(blur_img,127,255,cv2.THRESH_BINARY)
img4 = cv2.adaptiveThreshold(blur_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)


img = [img1,img2,img3,img4]
for i in range(0,4):
    show_img = cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB)
    plt.imshow(show_img)
    plt.show()
    
##結論: 先進行高斯濾波降噪，再進行高斯自適應圖像二值化效果比較好
```
