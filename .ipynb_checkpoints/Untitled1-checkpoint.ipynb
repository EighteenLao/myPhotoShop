```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Day 11 : 畫圓

```python
org_img = np.zeros((100,100,3),np.uint8)

circle_center = (50, 50)
radius = 30
color = (0, 255, 255)
thickness = 3


cv2.circle(org_img,circle_center, radius, color, thickness) #circle


cv2.circle(org_img,(80, 10), 10, (255, 0, 0), -1) #circle

cv2.circle(org_img,(10, 10), 1, (0, 255, 0), -1) #point


org_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB)
plt.imshow(org_img)
plt.show()
```

# Day12:畫矩形

```python
rg_img = np.zeros((100,100,3),np.uint8)

cv2.line(org_img, (0, 0), (70, 70), (0, 0, 255), 2)

cv2.rectangle(org_img, (20, 30), (60, 90), (0, 255, 0), 2) #左上座標、右下座標
cv2.rectangle(org_img, (70, 20), (80, 30), (255, 0, 0), -1) #左上座標、右下座標

cv2.ellipse(org_img, (50,50), (30,20), 45, 0, 360, (255,255,0), 2) #橢圓的長半徑&短半徑、傾斜角度、起始角度、終止角度
cv2.ellipse(org_img, (50,50), (30,20), -45, 0, 180, (0,255,255), 2)


org_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB)
plt.imshow(org_img)
plt.show()
```
