```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Day26: 銳化圖像(強化邊緣)

```python
def show_img(img):
    copy_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(copy_img)
    plt.show()
```

```python
src = cv2.imread('C:\\Users\\USER\\Pictures\\Wallpaper\\thumb-1920-946628.png')
```

```python
show_img(src)

blur_src = cv2.GaussianBlur(src, (0, 0),100)
dst = cv2.addWeighted(src, 1.5, blur_src, -0.5, 0)

show_img(dst)
```

# Day27: colorbar

```python
src = cv2.imread('C:\\Users\\USER\\Pictures\\Wallpaper\\thumb-1920-946628.png')
```

```python
result_src = cv2.applyColorMap(src, cv2.COLORMAP_AUTUMN)
show_img(result_src)

result_src = cv2.applyColorMap(src, cv2.COLORMAP_BONE)
show_img(result_src)

result_src = cv2.applyColorMap(src, cv2.COLORMAP_JET)
show_img(result_src)

result_src = cv2.applyColorMap(src, cv2.COLORMAP_RAINBOW)
show_img(result_src)
```
