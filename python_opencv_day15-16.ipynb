```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
src = cv2.imread('C:\\Users\\USER\\Pictures\\Wallpaper\\thumb-1920-946628.png') #絕對路徑使用'\'，相對路徑使用'/'
```

# Day15~16: 圖層比例分配、canny edge

# Day15: 圖層加減以及比例分配

```python
org_img = src 

mask = np.full(org_img.shape,128).astype(np.uint8)

img_add = cv2.add(org_img,mask)
img_sub = cv2.subtract(org_img,mask)
img_weight = cv2.addWeighted(org_img, 0.8, mask, 0.2, 0)

#
img_add = cv2.cvtColor(img_add, cv2.COLOR_BGR2RGB)
plt.imshow(img_add)
plt.show()
img_sub = cv2.cvtColor(img_sub, cv2.COLOR_BGR2RGB)
plt.imshow(img_sub)
plt.show()
img_weight = cv2.cvtColor(img_weight, cv2.COLOR_BGR2RGB)
plt.imshow(img_weight)
plt.show()
```

# Day16: 利用高斯模糊尋找邊緣

```python
org_img = src 

gray_img = cv2.cvtColor(org_img,cv2.COLOR_BGR2GRAY)

kernel_size = 2
blur_img = cv2.blur(gray_img, (kernel_size, kernel_size))

canny_img = cv2.Canny(blur_img, 30, 150)


blur_img = cv2.cvtColor(blur_img, cv2.COLOR_BGR2RGB)
plt.imshow(blur_img)
plt.show()
canny_img = cv2.cvtColor(canny_img, cv2.COLOR_BGR2RGB)
plt.imshow(canny_img)
plt.show()
```
