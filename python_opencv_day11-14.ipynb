# Day 11 : 畫圓

```python
org_img = np.zeros((100,100,3),np.uint8)

circle_center = (50, 50)
radius = 30
color = (0, 255, 255)
thickness = 3


cv2.circle(org_img,circle_center, radius, color, thickness) #circle
cv2.circle(org_img,(80, 10), 10, (255, 0, 0), -1) #circle
cv2.circle(org_img,(10, 10), 1, (0, 255, 0), -1) #point


org_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB)
plt.imshow(org_img)
plt.show()
```

# Day14: 分離以及合併RGB三通道

```python
#分離出三通道
org_img = src
(b,g,r) = cv2.split(org_img)

#無顏色值圖像
w,h,c = org_img.shape
zeros_color = np.zeros([w,h], dtype=np.uint8)

#合併三通道[b,g,r]
b_channel = cv2.merge([b, zeros_color, zeros_color])
g_channel = cv2.merge([zeros_color, g, zeros_color])
r_channel = cv2.merge([zeros_color, zeros_color, r])

#
b_channel = cv2.cvtColor(b_channel, cv2.COLOR_BGR2RGB)
plt.imshow(b_channel)
plt.show()

g_channel = cv2.cvtColor(g_channel, cv2.COLOR_BGR2RGB)
plt.imshow(g_channel)
plt.show()

r_channel = cv2.cvtColor(r_channel, cv2.COLOR_BGR2RGB)
plt.imshow(r_channel)
plt.show()
```

# 圖像中各顏色通道直方圖

```python
org_img = src
color = ('b','g','r')
plt.figure(figsize=(10,5))

#enumerate:索引以及矩陣裡的參數
for index, channel in enumerate(color):
    histogram = cv2.calcHist([org_img],[index],None,[256],[0, 256])
    plt.plot(histogram, color = channel)
    plt.xlim([0, 256])
```
