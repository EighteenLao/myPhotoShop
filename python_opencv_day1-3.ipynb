```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Day1、2、3 : 圖像的讀取、顯示、圖片裁切、旋轉和縮放

# 圖像的讀取、顯示

```python
src = cv2.imread('C:\\Users\\USER\\Pictures\\Wallpaper\\thumb-1920-946628.png') #絕對路徑使用'\'，相對路徑使用'/'
cv2.namedWindow("input image",cv2.WINDOW_AUTOSIZE)
cv2.imshow("input image",src)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
#opencv格式為BGR，所以要轉成RGB顯示
rgbsrc = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.show()
plt.imshow(rgbsrc)
```

# 圖片裁切、旋轉和縮放

```python
#裁切
org_img = rgbsrc
crop_src = org_img[100:500,300:750]
plt.imshow(crop_src)
plt.show()

#旋轉
(h,w,d) = org_img.shape
center = (w//2,h//2) #'//':取整數
M = cv2.getRotationMatrix2D(center,45,1)
rotate_src = cv2.warpAffine(org_img, M, (w, h)) #圖像仿射轉換
plt.imshow(rotate_src)
plt.show()

#縮放
w = int(org_img.shape[1]*1/2)
h = int(org_img.shape[0]*1/2)
resize_scr_size = (w,h) #size(width,height),different with image array
resize_scr = cv2.resize(org_img,resize_scr_size,interpolation = cv2.INTER_LINEAR) #INTER_LINEAR:鄰近像素差值
plt.imshow(resize_scr)
```
